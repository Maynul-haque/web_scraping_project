```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

api_name_series = []
api_category_series =[]
api_description_series =[]
page_number = 0
while page_number <897:
  url = 'https://www.programmableweb.com/category/all/apis?page={page_number}'
  r = requests.get(url)
  soup = BeautifulSoup(r.content, 'lxml')
  api_name_list = soup.find('div', class_ = 'main-container container').find('div', class_ = 'view-content').find_all('td', class_ ='views-field views-field-pw-version-title')
  api_name = [item.get_text() for item in api_name_list]
  for item in api_name:
    api_name_series.append(item)
  api_category_list = soup.find('div', class_ = 'main-container container').find_all('td', class_ ='views-field views-field-field-article-primary-category')
  api_category = [item.get_text() for item in api_category_list]
  for item in api_category:
    api_category_series.append(item)
  api_description_list = soup.find('div', class_ = 'main-container container').find('div', class_ ='view-content').find_all('td', class_ ='views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8')
  api_description = [item.get_text() for item in api_description_list]
  for item in api_description:
    api_description_series.append(item)
  page_number += 1

data = {'API Name': api_name_series, 'Category': api_category_series, 'Description': api_description_series}
df = pd.DataFrame(data)
df.to_csv('Api scraping.csv')
```
